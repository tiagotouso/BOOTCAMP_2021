```python
import numpy as np
import pandas as pd
import random as rd
```

# DATASET EXEMPLO I

```python
tamanho = 20
rd.seed = 42

sx = [rd.sample(['F', 'M'], 1)[0] for _ in range(tamanho)]
idade = [rd.sample(list(range(18, 75)), 1)[0] for _ in range(tamanho)]
classe = [rd.sample(['A', 'B', 'C'], 1)[0] for _ in range(tamanho)]

dic = {'sexo': sx, 'idade': idade, 'classe': classe}

dados = pd.DataFrame(dic)
dados.sample(5)
```

# * LABELENCODER

```python
from sklearn.preprocessing import LabelEncoder
```

```python
tf = LabelEncoder()
tf.fit(dados[['classe']])
```

```python
resultado = tf.fit_transform(dados[['classe']])
resultado
```

```python
tf.inverse_transform(resultado)
```

# * ONEHOTENCODER

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
tf = OneHotEncoder()
tf.fit(dados[['classe']])
```

```python
resultado = tf.fit_transform(dados[['classe']]).toarray()
resultado
```

```python
tf.inverse_transform(resultado)
```

# * MINMAXSCALER

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
tf = MinMaxScaler()
tf.fit(dados[['idade']])
```

```python
resultado = tf.fit_transform(dados[['idade']])
resultado
```

```python
tf.inverse_transform(resultado)
```

# * STANDARDSCALER

```python
from sklearn.preprocessing import StandardScaler
```

```python
tf = StandardScaler()
tf.fit(dados[['idade']])
```

```python
resultado = tf.fit_transform(dados[['idade']])
resultado
```

```python
tf.inverse_transform(resultado)
```

# DATASET EXEMPLO II

```python
dados = pd.DataFrame({'cores':['Amarelo', 'Verde', 'vermelho'], 'formas': ['circulo', 'quadrado', 'triângulo']})
dados
```

# * GET_DUMMIES

```python
pd.get_dummies(dados)
```

# * ONEHOTENCODER

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
tf = OneHotEncoder(handle_unknown='ignore', sparse=False)
tf.fit(dados)
```

```python
tf.categories_
```

```python
resultado = tf.fit_transform(dados)
resultado
```

```python
tf.feature_names_in_
```

```python
tf.get_feature_names_out()
```

```python
tf.inverse_transform(resultado)
```

# * ORDINALENCODER

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
tf = OrdinalEncoder()
tf.fit(dados)
```

```python
tf.categories_
```

```python
resultado = tf.fit_transform(dados)
resultado
```

```python
dadosnew = pd.DataFrame(tf.fit_transform(dados), columns=dados.columns)
dadosnew
```

```python
tf.inverse_transform(dadosnew)
```
